```python
import torch
from torch import nn, optim

class Model(nn.Module):
    def __init__(self):
        super().__init__()
        self.l1 = nn.Linear(10,20)
        self.relu = nn.ReLU()
        self.l2 = nn.Linear(20, 5)
    def forward(self, x):
        x = self.l1(x)
        x = self.relu(x)
        x = self.l2(x)
        return x
X = torch.rand(100, 10)
Yd = torch.rand(100, 5)

model = Model()
loss_f = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)
epochs = 100
for ep in range(epochs):
    Y = model(X)
    optimizer.zero_grad()
    loss = loss_f(Y, Yd)
    loss.backward()
    optimizer.step()
    if ep % 10 == 0:
        print(f"{ep}: {loss:.2e}")
        
print(model.l1.weight, model.l1.bias, model.l2.weight, model.l2.bias)
```
